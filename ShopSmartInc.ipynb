```python
#import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
# utility function to print markdown string
def printmd(string):
    display(Markdown(string))
```

```python
#load the customers data. This contains information concerning the customer_id, device_id, location, currency
customers = pd.read_csv('customers.csv')
```

```python
customers
```

```python
#load the events data. This contains information concerning the event_id, customer_id, event_data and event_timestamp
events = pd.read_csv('events.csv')
events
```

```python
#load the products data. This contains information concerning id, product name and price
products = pd.read_csv('products.csv')
products
```

```python
#load the orders data. This contains information concerning the order_id, customer_id, order status, time the customer checked out
orders= pd.read_csv('orders.csv')
orders
```

```python
#load the line_items data. This contains information concerning the line_item_id, order_id, item_id, quantity
line_items = pd.read_csv('line_items.csv')
line_items
```

```python
#Merges the events and customers dataframes using a left join. Here, the number of rows of the output dataframe is the same as the no of rows as the left dataframe
df1 = pd.merge(events,customers[['customer_id','device_id','location','currency']], on = 'customer_id', how = 'left')
```

```python
df1
```

```python
#Merges the df1 and orders dataframes using a left join. Here, the number of rows of the output dataframe is the same as the no of rows as the left dataframe
df2 = pd.merge(df1,orders[['order_id','customer_id','status','checked_out_at']], on = 'customer_id', how = 'left')
```

```python
df2
```

```python
#Merges the line_items and orders dataframes using a left join. Here, the number of rows of the output dataframe is the same as the no of rows as the left dataframe
df3 = pd.merge(line_items,orders[['order_id','customer_id','status','checked_out_at']], on = 'order_id', how = 'left')
```

```python
df3
```

```python
#Merges the df3 and products dataframes using an inner join. All common items having the same item_id in the left dataframe and id in the right dataframe are replicated in the new dataframe.
df4 = pd.merge(df3,products[['id','name','price']], left_on = 'item_id', right_on = 'id', how = 'inner')
```

```python
df4
```

```python
#Merges the df2 and df4 dataframes using an inner join. All common items having the same item_id in the left dataframe and id in the right dataframe are replicated in the new dataframe
df5 = pd.merge(df2,df4[['order_id','customer_id','status','checked_out_at','line_item_id','item_id','quantity','id','name','price']], on = 'customer_id', how = 'inner')
```

```python
df5
```

```python
#Drop column duplicates
df5 = df5.drop(['order_id_y','status_y','checked_out_at_y','id'],axis=1)
```

Feature Engineering of the Events Data

```python
#Upon closer inspection, events_data is a dictionary-like object containing item_id, quantity, timestamp, event_type, status and order_id, all of which are 
    #present in our dataframe, except event_type. So we'll extract the event type from the event_data column

#import json library to handle json-like dictionary
import json
#Create an empty list to store the extracted event types
event_types = []

# Iterate over each row in the DataFrame
for idx, row in df5.iterrows():
    try:
        # Load the JSON string into a dictionary
        data_dict = json.loads(row['event_data'])
        # Extract the event_type
        event_type = data_dict.get('event_type')
        # Append the extracted event_type to the list
        event_types.append(event_type)
    except (json.JSONDecodeError, TypeError, KeyError):
        # Handle invalid JSON strings or missing 'event_type' key
        event_types.append(None)

# Add the list of event types as a new column to the DataFrame
df5['event_type'] = event_types
#Drop event_data column
df5.drop('event_data', axis = 1,inplace=True)
```

```python
df5.head()
```

```python
#Extract day, month and year columns (from the Event_timestamp column) for each of the transactions on the platform using the Datetime Module.
#Extract date from the event_timestamp data
df5['event_date'] = df5.event_timestamp.str.split('.').str.get(0)
#Import datetime module
from datetime import datetime
event_date = df5.event_date

event_date = pd.to_datetime(event_date, format='%Y-%m-%d %H:%M:%S', errors='coerce')

df5['event_date'] = event_date

```

```python

#Create new columns
df5['event_year'] = df5['event_date'].dt.year
df5['event_month'] = df5['event_date'].dt.strftime('%B')
df5['event_day'] = df5['event_date'].dt.strftime('%A')
```

```python
def get_time_of_day (dt):
    hour = dt.hour
    if hour in range(0, 12):
        return "morning"
    elif hour in range(12, 18):
        return "afternoon"
    else:
        return "night"


df5['event_time_of_day'] = df5 ['event_date'].apply(get_time_of_day)
```

```python
import matplotlib.pyplot as plt

# Calculate visit counts for each customer
visit_counts = visit_df.groupby('customer_id').size()

# Sort customers based on visit frequency
visit_counts_sorted = visit_counts.sort_values(ascending=False)

# top 20 customers
top_n = 20 
top_n_customers = visit_counts_sorted.head(top_n)

# Plot bar chart of top N customers
plt.bar(top_n_customers.index, top_n_customers.values, color='skyblue')
plt.xlabel('Customer ID')
plt.ylabel('Visit Count')
plt.title(f'Top {top_n} Customers by Visit Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y')
plt.show()
```

```python
from IPython.display import Markdown
printmd('#### These are the top 20 customers who visited ShopSmart Inc, with the highest number of visits being 120')
```

```python

#Daily Visits
# Group by date and count visits
daily_visits = visit_df.groupby(df5['event_date'].dt.date).size()

# Plot line chart
plt.figure(figsize=(10, 6))
daily_visits.plot(kind='line', marker='o', color='b')
plt.title('Daily Visits')
plt.xlabel('Date')
plt.ylabel('Number of Visits')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

printmd('#### Visits peaked during the last days in March, corresponding to the 30th and 31st. Users were probably trying to get some weekend shopping done. In comparison with 23rd and 24th , also a weekend but with little or no visits, I can assume that ShopSmart Inc probably had a discount sale during the weekend of the 30th')
```

```python
# Group by week and count visits
weekly_visits = visit_df.groupby(pd.Grouper(key='event_date', freq='W')).size()

# Plot line chart for weekly visits
plt.figure(figsize=(10, 6))
weekly_visits.plot(kind='line', marker='o', color='b')
plt.title('Weekly Visits')
plt.xlabel('Week')
plt.ylabel('Number of Visits')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

printmd('#### Visits grew steadily in the last week of March, reaching maximum on the 31st of the March')
```

```python
# Group by month and count visits
monthly_visits = visit_df.groupby(pd.Grouper(key='event_date', freq='M')).size()

# Plot line chart for monthly visits
plt.figure(figsize=(10, 6))
monthly_visits.plot(kind='line', marker='o', color='b')
plt.title('Monthly Visits')
plt.xlabel('Month')
plt.ylabel('Number of Visits')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

printmd('#### From this chart, it is obvious that the provided data only contains shoppers activity for the month of March')
```

```python
#group customers by country
customer_counts = df6.groupby('location').size().reset_index(name='customer_id')
```

```python
customer_counts.reset_index(drop=True, inplace=True)
customer_counts.index = customer_counts.index + 1
```

```python
customer_counts = customer_counts.sort_values(by='customer_id', ascending=False)
customer_counts
```

```python
#This is a map of customers who actually purchased items from the website
mymap = folium.Map(location=[0, 0], tiles = 'Cartodbpositron', zoom_start=3.5)

# Add markers for each country
for index, row in customer_counts.iterrows():
    location = row['location']
    count = row['customer_id']
    # Read the world_csv file containing the latitude and longitudes of countries
    centroid = pd.read_csv("world_data.csv")
    centroid = centroid[centroid['location'] == location].iloc[0]
    folium.Marker(location=[centroid['latitude'], centroid['longitude']],
                  popup=f'{location}: {count} customers',
                  icon=folium.Icon(color='blue', icon='info-sign')).add_to(mymap)

# Display map
mymap

```

```python
printmd('#### From the map above, it appears that there was no customer purchase from Russia, Libya and the United States')
```

```python
#What products was ordered most per country
product_count_by_country = df6.groupby(['location', 'name']).size().reset_index(name='quantity')
most_ordered_product_by_country = product_count_by_country.loc[product_count_by_country.groupby('location')['quantity'].idxmax()]
```

```python
#most_ordered_product_by_country
# Create a base map
mymap2 = folium.Map(location=[0, 0],tiles = 'Cartodbpositron', zoom_start=3.5)

# Add markers for each country
for index, row in most_ordered_product_by_country.iterrows():
    country = row['location']
    product = row['name']
    quantity = row['quantity']
    # Read the world_csv file containing the latitude and longitudes of countries
    centroid = pd.read_csv("world_data.csv")
    centroid = centroid[centroid['location'] == country].iloc[0]
    folium.Marker(location=[centroid['latitude'], centroid['longitude']],
                  popup=f'{country}: {product} ({quantity} orders)',
                  icon=folium.Icon(color='blue', icon='info-sign')).add_to(mymap2)
# Display map
mymap2
```

```python
printmd('#### The country with the most purchases Mauritius ordered more of Dell XPS 13 Laptop than any other product. In subharan Africa, tech products were preferred by consumers, in the following order; Nigeria (55 Inch 4K smart TV), Benin, Togo (Samsung Galaxy S21), Ghana and Equitorial Guinea (iPhone 13); Cote dIvore and Liberia (Dell XPS 13); Gabon (Apple Airpods Pro); Guinea and Niger (Fitbit Charger 4); Cameroon, Mali, Sierra Leone, Gambia and Senegal (Adidas Ultraboost), Chad (Microsoft Xbox Series X) and Guinea Bissau (SONY PlayStation 5). This preference for tech products by subsaharan countries means that they could be buying their non-tech products from another store. To increase patronage, ShopSmart Inc should do a discount sale on non-tech products, multiple discounts on multiple purchases.')
```

```python
#How does time of day affect purchases

# Create a base map
mymap3 = folium.Map(location=[0, 0], tiles='Cartodbpositron', zoom_start=3.5)

# Calculate time_day using value_counts
time_day = df6.groupby('location')['event_time_of_day'].apply(lambda x: x.value_counts().idxmax())

# Color differences
def color_producer(x):
    if x == 'morning':
        return 'forestgreen'
    elif x == 'afternoon':
        return 'darkred'
    else:
        return 'black'

# Add Bubble Map
for location, time_of_day in time_day.items():
    lat, lon = df6.loc[df6['location'] == location, ['latitude', 'longitude']].iloc[0]
    folium.Circle(location=[lat, lon], radius=2000, color=color_producer(time_of_day)).add_to(mymap3)

mymap3
```

```python
printmd('#### From the map above, all purchases were made in the morning. ShopSmart Inc further increase sales by posting their best deals between the hours of 00 and 11am')
```

```python

# Create a base map
mymap4 = folium.Map(location=[0, 0], tiles='Cartodbpositron', zoom_start=3.5)

# Calculate day using value_counts
day = df6.groupby('location')['event_day'].apply(lambda x: x.value_counts().idxmax())

# Color differences
def color_producer(x):
    if x == 'Sunday':
        return 'forestgreen'
    elif x == 'Monday':
        return 'darkred'
    elif x == 'Tuesday':
        return 'red'
    elif x == 'Wednesday':
        return 'brown'
    elif x == 'Thursday':
        return 'blue'
    elif x == 'Friday':
        return 'yellow'
    elif x == 'Saturday':
        return 'purple'
    else:
        return 'black'

# Add Bubble Map
for location, day_of_week in day.items():
    lat, lon = df6.loc[df6['location'] == location, ['latitude', 'longitude']].iloc[0]
    folium.Circle(location=[lat, lon], radius=2000, color=color_producer(day_of_week)).add_to(mymap4)

mymap4
```

```python
printmd('#### Most purchases were made on Saturday')
```
